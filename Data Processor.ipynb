```python
from os import listdir
from matplotlib import image
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import pandas as pd 
import numpy as np
```

```python
def to_pixel(path,width,height):
    color_img = Image.open(path)
    color_img = color_img.resize((width, height), Image.ANTIALIAS)
    gray_img = ImageOps.grayscale(color_img)
    return np.asarray(gray_img).flatten().astype(int).tolist()
    
```

```python
#Create the dataset

def create_dataset(width,height):
    images = []
    for j in ['test/', 'train/']:
        path1 = "data_upload_v2/" + j
        for i in ["covid/", "non/"]:
            path2 = path1+i
            if j == 'test/' and i== "non/":
                for k  in listdir(path2):
                    path3 = path2+k
                    for l in listdir(path3):
                        images.append([path3 + '/' + l] + to_pixel(path3 + '/' + l,width,height) + [0])
            else:
                for k  in listdir(path2):
                    path3 = path2+k
                    if i == "non/":
                        images.append([path3] + to_pixel(path3,width,height) + [0])
                    else:
                        images.append([path3] + to_pixel(path3,width,height) + [1])
    return images, width*height
 
```

```python
dataset = create_dataset(149,118)
data = dataset[0]
num_features = dataset[1]
paths = pd.DataFrame(data = data, columns = ["filename"]+['pixel'+ str(i) for i in range(num_features)]+["Tested"])
```
